```python
#Import the required Libraries
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt  
```

```python
#Load the Dataset
data=pd.read_csv("/content/SampleSuperstore.csv")
```

```python
#Check for Null values
if(data.isnull==True):
  print("Null data is there")
else:
  print("No null data")
```

```python
#Print the first five rows
data.head()
```

```python
#Shape of data - (rows,columns)
data.shape
```

```python
#Checking is there any categories in country 
data['Country'].value_counts()
```

**There is only one country - United States**
**so we can eliminate this column without any interupt.**




```python

#dropping Country column
data=data.drop('Country',axis=1)
```

```python
data.head()
```

```python
#country column is removed from the dataset
data.shape
```

```python
#Unique Shipments in the dataset and their counts
data['Ship Mode'].unique()
data['Ship Mode'].value_counts()
```

```python
#Unquie segment in the data and their counts
data['Segment'].unique()
data['Segment'].value_counts()
```

```python
#Unique Region in the dataset and their counts
data['Region'].unique()
data['Region'].value_counts()
```

```python
#Unquie Category in the dataset and their counts
data['Category'].unique()
data['Category'].value_counts()
```

```python
#Visualsing The Categories in a pie chart
plt.figure(figsize=(10,8))
data['Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
#Unique Subcategories and their counts
data['Sub-Category'].unique()
data['Sub-Category'].value_counts()
```

```python
#Visualising the Sub-Categories in a pie chart
plt.figure(figsize=(10,8))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

**Profits in Each Customer Segment**

```python

data.groupby('Segment')['Profit'].agg(['sum']).sort_values(['sum'], 
                                                         ascending=False).plot.bar(figsize=(10,6));
```

Consumer Segment gives the major profit whereas Home office giving the least.

**Profits in Each Region**

```python
data.groupby('Region')['Profit'].agg(['sum']).sort_values(['sum'], 
                                                         ascending=False).plot.bar(figsize=(10,6));
```

West region has the highest profift with Central region having the least.

**Profits in Each Category**

```python
data.groupby('Category')['Profit'].agg(['sum']).sort_values(['sum'], 
                                                         ascending=False).plot.bar(figsize=(10,6));
```

Technology has the highest profit whereas Furniture has the least profit.

**Profits in Sub-Category**

```python

data.groupby('Sub-Category')['Profit'].agg(['sum']).sort_values(['sum'], 
                                                         ascending=False).plot.bar(figsize=(16,10));
```

Highest profit is earned in Copiers and the least earned in tables.

**Profits in Each Shipping mode**

```python
data.groupby('Ship Mode')['Profit'].agg(['sum']).sort_values(['sum'], 
                                                         ascending=False).plot.bar(figsize=(10,6));
```

Standard Class has earned the highest profit with same Day earning the least.

**Sales Done in Each Sub-Category**

```python
plt.figure(figsize=(8,6))
data.groupby('Sub-Category')['Sales'].agg(['sum']).plot.bar(figsize=(16,10))
```

```python

#Count of Sub-Category region wise
plt.figure(figsize=(20,10))
sns.countplot(x="Sub-Category", hue="Region", data=data)
plt.show()
```
